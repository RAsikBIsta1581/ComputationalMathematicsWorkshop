2. Write a Python function that takes a number as a parameter and checks whether the number
is prime or not.

```python
def is_prime(n):
  if n <= 1:
        return False
  if n == 2:
        return True
  if n % 2 == 0:
        return False
  for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
  return True
print(is_prime(2))
print(is_prime(15))
print(is_prime(17))
```

3. Write a program to print twin primes less than 1000. If two consecutive odd numbers are
both prime then they are known as twin primes.

```python
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

def twin_primes(limit):
    for i in range(3, limit, 2):
        if is_prime(i) and is_prime(i + 2):
            print(f"({i}, {i+2})")

twin_primes(1000)
```

4. A positive integer is called a perfect number if it is equal to the sum of all of its divisors,
including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a
function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors
(excluding n).

```python
def sumOfDivisors(n):

    if n <= 0:
        raise ValueError("Input must be a positive integer")

    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

print(sumOfDivisors(6))
print(sumOfDivisors(28))
print(sumOfDivisors(12))
```

6. Write the python program to find the mean median from the following list:
Marks = [20, 30, 35, 55, 70, 80, 90] Do not use the library function if there is one.

```python
Marks = [20, 30, 35, 55, 70, 80, 90]

total = 0
for i in Marks:
    total += i

mean = total / len(Marks)

n = len(Marks)

if n % 2 == 1:
    median = Marks[n // 2]
else:
    median = (Marks[n // 2 - 1] + Marks[n // 2]) / 2

print("Mean =", mean)
print("Median =", median)
```

7. Write a function that converts a decimal number to binary number.

```python
def decimal_to_binary(n):
    if n == 0:
        return "0"

    binary = ""

    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2

    return binary

num = 1581
print("Binary of", num, "is", decimal_to_binary(num))
```

8. Write a function prodDigits() that inputs a number and returns the product of digits of that
number.

```python
def prodDigits(n):
    product = 1

    while n > 0:
        digit = n % 10
        product *= digit
        n = n // 10

    return product

num = 1581
print("Product of digits:", prodDigits(num))
```

9. Write a function that will take in an unknown number of arguments and multiply all of them together and run the function for these 2 sets of numbers:

a. 1,2,3,4,5

b. 12,13,14

```python
def multiply_all(*args):
    result = 1
    for num in args:
        result *= num
    return result

# a) 1, 2, 3, 4, 5
print("Product (a):", multiply_all(1, 2, 3, 4, 5))

# b) 12, 13, 14
print("Product (b):", multiply_all(12, 13, 14))
```

**Example 3:** Perform the following operations on given sets.

set1 = {10, 20, 30, 40, 50}

a. Find the length of set1

c. Union

d. Intersection

m. Print the maximum and minimum value in set 2

```python
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set3 = {10, 20, 30}

# a. Length of set1
print("a. Length of set1:", len(set1))

# b. Sum of elements in set2
print("b. Sum of set2:", sum(set2))

# c. Union
print("c. Union:", set1 | set2)

# d. Intersection
print("d. Intersection:", set1 & set2)

# e. set1 - set2
print("e. set1 - set2:", set1 - set2)

# f. set2 - set1
print("f. set2 - set1:", set2 - set1)

# g. Check subset
print("g. set3 subset of set1:", set3.issubset(set1))
print("   set3 subset of set2:", set3.issubset(set2))

# h. Complement of set1 (U = 1 to 50)
U = set(range(1, 51))
print("h. Complement of set1:", U - set1)

# i. Check if 15 is present in set1
print("i. Is 15 in set1?", 15 in set1)

# j. Add 60 to set1
set1.add(60)
print("j. After adding 60 to set1:", set1)

# k. Remove 60 from set2
set2.discard(60)
print("k. After removing 60 from set2:", set2)

# l. Remove all elements from set3
set3.clear()
print("l. After clearing set3:", set3)

# m. Max and Min of set2
print("m. Max of set2:", max(set2))
print("   Min of set2:", min(set2))
```
