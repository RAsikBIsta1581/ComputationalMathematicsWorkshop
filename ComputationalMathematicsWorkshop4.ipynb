**Scatter:**

Scatter graphs will plot one point per observation, we can use these to graph independent
variables to see if there is a correlation. Python will require two sets of variables for a scatter
graph, one for the x-axis and another for the corresponding y axis. These sets will need to be of
the same length and work best as lists, array, or array-like data types. We can call ‘pyplot’ to plot
a scatter with the pyplot.scatter(x,y) call for example.

**Task 1:**
Create a scatter graph of 10 random x and y values between 0 and 50. Label the axis to show:

x: Time taken in seconds

y: Distance reached

Title “Buzzer Game Results”


One you have a working program, add the following code to your program;
colors=["red","green","blue","yellow","pink","orange","purple","brown"
,"cyan","magenta"]

And add to your scatter() call the ‘c=colors’ argument like so;

plt.scatter(x, y, c=colors)

What happens?

```python
import matplotlib.pyplot as plt
import random

x = [random.randint(0, 50) for _ in range(10)]
y = [random.randint(0, 50) for _ in range(10)]

colors = ["red", "green", "blue", "yellow", "pink",
          "orange", "purple", "brown", "cyan", "magenta"]

plt.scatter(x, y, c=colors)

plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")

plt.show()
```

**Ans:** Each point in the user's scatter plot gets a different color from he list. Since the user have exactly 10 points and 10 colors, each point is uniquely colored which makes the plot more visually engaging and helps to distinguish individual data points.

**Task 2:**
Following data shows the change in population of different animals from 2017 to 2022.

a. Create a pie chart for the population of each year.

b. Can we show all the pie chart on the single pane? Write the code.

Screenshot 2026-04-01 101015.png

```python
import matplotlib.pyplot as plt

years = [2017, 2018, 2019, 2020, 2021, 2022]
bears = [8, 54, 93, 116, 137, 184]
dolphins = [150, 77, 32, 1, 6, 1]
whales = [80, 54, 100, 76, 93, 72]

animals = ["Bears", "Dolphins", "Whales"]

# ---a. Create a pie chart for the population of each year. ---
plt.pie([bears[0], dolphins[0], whales[0]], labels=animals, autopct='%1.1f%%')
plt.title("Animal Population in 2017")
plt.show()

# ---b. Can we show all te pie chart on the single pane? Write the code. ---
fig, axes = plt.subplots(2, 3, figsize=(12, 8))

for i, year in enumerate(years):
    row = i // 3
    col = i % 3
    axes[row, col].pie([bears[i], dolphins[i], whales[i]], labels=animals, autopct='%1.1f%%')
    axes[row, col].set_title(f"Population in {year}")

plt.tight_layout()
plt.show()
```

**Task 3:** Construct a pie chart to visually display the favorite fruits of the students in a class based on the given data:

Mango - 45; Orange - 30; Plum - 15; Pineapple - 30; Melon – 30

```python
import matplotlib.pyplot as plt

fruits = ["Mango", "Orange", "Plum", "Pineapple", "Melon"]
counts = [45, 30, 15, 30, 30]

plt.pie(counts, labels=fruits, autopct='%1.1f%%', startangle=90)

plt.title("Favorite Fruits of Students")

plt.show()
```
