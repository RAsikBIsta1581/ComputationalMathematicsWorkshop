**Task 2:**

The following table shows the different vehicles and the distance covered by them in the span of 5 days. Display the following table in line plot.

a. Plot “days” in X- axis and “distance covered” in Y- axis.

b. Title of the table “bike details in Plot”

Screenshot 2026-03-25 104950.png

```python
import matplotlib.pyplot as plt

# Days on X-axis
days = ["Day 1", "Day 2", "Day 3", "Day 4", "Day 5"]

# Distances for each bike
enfield = [50, 40, 70, 80, 20]
honda   = [80, 20, 20, 50, 60]
yahama  = [70, 20, 60, 40, 60]
ktm     = [40, 20, 30, 70, 60]

# Plot each bike’s data
plt.plot(days, enfield, marker='o', label="Enfield")
plt.plot(days, honda, marker='o', label="Honda")
plt.plot(days, yahama, marker='o', label="Yahama")
plt.plot(days, ktm, marker='o', label="KTM")

# Labels and title
plt.xlabel("Days")
plt.ylabel("Distance Covered (kms)")
plt.title("Bike details in Plot")

# Add grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
