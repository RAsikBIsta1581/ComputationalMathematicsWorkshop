a. Create three vectors in 3D space and plot them using Matplotlib.

U= np.array([1, 0.25 ,0])

V = np.array([0, 1, 0.25])

W = np.array([0.25, 0, 1])

To do 3D plot, you will need to create 3D axes using following code:

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

```python
import numpy as np
import matplotlib.pyplot as plt
U = np.array([1, 0.25, 0])
V = np.array([0, 1, 0.25])
W = np.array([0.25, 0, 1])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(0, 0, 0, U[0], U[1], U[2], color='r', label='U')
ax.quiver(0, 0, 0, V[0], V[1], V[2], color='g', label='V')
ax.quiver(0, 0, 0, W[0], W[1], W[2], color='b', label='W')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.legend()
plt.title("3D Vectors Plot")
plt.show()
```

b. Create two vectors in 2D space, and add them to create a third vector, and plot all three vectors using Matplotlib.

U = np.array([-0.5, 0.5])

V = np.array([1,1])

The two vectors are added as:

 W = U + V # adding two vector

```python
import numpy as np
import matplotlib.pyplot as plt
U = np.array([-0.5, 0.5])
V = np.array([1, 1])
W = U + V
plt.figure()
ax = plt.gca()
ax.quiver(0, 0, U[0], U[1], angles='xy', scale_units='xy', scale=1, color='r', label='U')
ax.quiver(0, 0, V[0], V[1], angles='xy', scale_units='xy', scale=1, color='g', label='V')
ax.quiver(0, 0, W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b', label='W = U + V')
plt.xlim(-1, 2)
plt.ylim(-1, 2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Vector Addition")
plt.grid(True)
plt.legend()
plt.show()
```

c. Write a function to calculate magnitude of a vector. For example, calculate the magnitude of the vector be U = (1.5, 1.5, 2.0)

```python
import numpy as np
def vector_magnitude(vec):
    return np.linalg.norm(vec)
U = np.array([1.5, 1.5, 2.0])
magnitude = vector_magnitude(U)
print("Magnitude of U:", magnitude)
```

d. Write a function to calculate the direction cosines of the vector. For example, let the vector be O = (1.5, 1.5, 2.0).

```python
import numpy as np
def direction_cosines(vec):
    magnitude = np.linalg.norm(vec)
    return vec / magnitude
O = np.array([1.5, 1.5, 2.0])
cosines = direction_cosines(O)
print("Direction cosines:", cosines)
```

e. Write a function to calculate Scalar (dot) product of two vectors. For example, let the vectors be a= (1.5, 1.5, 2.0) and b = (−1.5, 0.5, 2.0).

```python
import numpy as np
def dot_product(vec1, vec2):
    return np.dot(vec1, vec2)
a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
result = dot_product(a, b)
print("Dot product of a and b:", result)
```

f. Write a function to calculate Vector (cross) product of two vectors. For example, let the vectors be a = (1.5, 1.5, 2.0) and b= (−1.5, 0.5, 2.0).

```python
import numpy as np
def cross_product(vec1, vec2):
    return np.cross(vec1, vec2)
a = np.array([1.5, 1.5, 2.0])
b = np.array([-1.5, 0.5, 2.0])
result = cross_product(a, b)
print("Cross product of a and b:", result)
```
