```python
import numpy as np
import matplotlib.pyplot as plt
""" Part 1 - Problem 1: Solve linear systems graphically. """
# Generate 100 equidistant point between -5 to 5 for x
x = np.linspace(-5,5,100)
## Calculate y values for Equation 1 and 2
y1 = (5-x)/3
y2 = (5+x)/2
y3 = 3-x
## Generate a figure frame for plotting the lines
fig, ax = plt.subplots()
## Plot the two lines
ax.plot(x,y1, color='r')
ax.plot(x,y2, color='g')
ax.plot(x,y3, color='b')
## To decorate the plot, we insert equations.
ax.text(0.5,0.6, '$x+3y = 5$', color='r')
ax.text(-3,0.5,'$x-2y = -5$', color='g')
ax.text(1.2,2, '$x+y = 3$', color='b')
## Set the limits of the X & Y axes.
ax.set_xlim(-4,4)
ax.set_ylim(-2,6)
## Set property of X and Y axes, e.g., color and linewidth
ax.axvline(color='k',linewidth = 1)
ax.axhline(color='k',linewidth = 1)
## This option specifies the ticks based the list of numbers provided.
ax.set_xticks(list(range(-4,5)))
ax.set_aspect('equal')
ax.grid(True,ls=':')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

**Question:** Write the conclusion from the above graph.

**Answer:** The given graph illustrates three linear equations plotted on the same coordinate plane and they are: x + y = 3, x + 3y = 5 and x - 2y = -5. Each line intersects with the others at distinct points which represents the solutions to pairs of equations. However, the three lines do not meet at a single common point, this means there is no solution that satisfies all three equations simultaneously. This shows that while each pair of equations has a unique solution, the overall system of three equations is inconsistent. In the graph, the plot highlights how the intersections reveal solutions to the subsystems but also how the absence of a single intersection point indicates that the full system has no common solution.

**Task 1:**

Plot the graph for linear system b and write the point of intersection between three lines.

```python
import numpy as np
import matplotlib.pyplot as plt
# Define x range
x = np.linspace(-5, 5, 100)
# Equations
y1 = (5 - x) / 3       # x + 3y = 5
y2 = (x + 5) / 2       # x - 2y = -5
y3 = 3 - x             # x + y = 3
# Plot lines
fig, ax = plt.subplots()
ax.plot(x, y1, color='r', label='$x+3y=5$')
ax.plot(x, y2, color='g', label='$x-2y=-5$')
ax.plot(x, y3, color='b', label='$x+y=3$')
# Intersection points
points = [(2, 1), (1/3, 8/3), (-1, 2)]
labels = ['(2,1)', '(1/3,8/3)', '(-1,2)']
for (px, py), label in zip(points, labels):
    ax.plot(px, py, 'ko')  # black dot
    ax.text(px+0.1, py, label, fontsize=9)
# Decorations
ax.axhline(0, color='k', linewidth=1)
ax.axvline(0, color='k', linewidth=1)
ax.set_xlim(-4, 4)
ax.set_ylim(-2, 6)
ax.set_xticks(range(-4, 5))
ax.set_yticks(range(-2, 7))
ax.set_aspect('equal')
ax.grid(True, ls=':')
ax.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear System b: Intersection Points")
plt.show()
```

**Numpy.linalg:**

linalg is a submodule in NumPy, a popular Python library used for numerical computations, particularly in the domain of linear algebra. linalg stands for "linear algebra". This submodule provides functions and classes for various linear algebra operations, such as matrix decompositions, eigenvalue computations, matrix multiplication, solving linear equations, and many others.

**Some useful linalg functions:**

1. linalg.det(a): calculates the determinant of the matrix.
2. linalg.matrix_rank(a): computes the rank of the matrix.
3. linalg.inv(a): calculates the multiplicative inverse of a matrix.
4. linalg.solve(a,b): solve a linear system of equations ax=b.
5. linalg.pinv(a): compute the pseudo-inverse of a matrix.

1. Create a matrix 𝑨 of order 5 × 5 with random entries. Calculate the rank of the matrix 𝑨 to check the number of linearly independent columns or rows and to check if the matrix is invertible. What are the other ways to check if the matrix is invertible?

```python
import numpy as np
A = np.random.randint(1, 21, (5, 5))
print("Matrix A (5x5):")
print(A)
rank = np.linalg.matrix_rank(A)
print("\nRank of Matrix A:", rank)
if rank == A.shape[0]:
    print("Matrix A is invertible (full rank).")
else:
    print("Matrix A is NOT invertible (rank deficient).")
```

Other than checking the rank, there are several ways to determine if a matrix is invertible. The most direct method is computing the determinant: if it is nonzero, the matrix is invertible; if it is zero, the matrix is singular. Another approach is examining the eigenvalues: a matrix is invertible if none of its eigenvalues are zero. You can also check the condition number; a finite and reasonably small condition number indicates invertibility, while a very large or infinite value suggests the matrix is nearly singular or non-invertible. Finally, attempting to compute the inverse directly (using np.linalg.inv) will either succeed if the matrix is invertible or raise an error if it is not. These methods provide complementary ways to assess invertibility beyond rank.

2. Generate a matrix 𝑨 of order 5 × 5 with random entries and generate a vector (column) 𝒙 of length 5 with random entries. Calculate Matrix-vector product 𝒃 = 𝑨𝒙. Now, we have a linear system, i.e. the matrix 𝑨 and a vector 𝒃. Solve this Linear System.

```python
import numpy as np
A = np.random.randint(1, 21, (5, 5))
x = np.random.randint(1, 21, (5, 1))
b = np.dot(A, x)
x_solution = np.linalg.solve(A, b)
print("Matrix A (5x5):")
print(A)
print("\nVector x (length 5):")
print(x)
print("\nVector b = A * x:")
print(b)
print("\nSolution of the Linear System (x):")
print(x_solution)
```

3. Solve the following system.

    8x+3y−2z=9

    −4x+7y+5z==15

    3x+4y−12z=35

```python
import numpy as np

# Step 1: Define the coefficient matrix A
A = np.array([[8, 3, -2],
              [-4, 7, 5],
              [3, 4, -12]])

# Step 2: Define the constants vector b
b = np.array([9, 15, 35])

# Step 3: Solve the system Ax = b
solution = np.linalg.solve(A, b)
print("Coefficient Matrix A:")
print(A)
print("\nConstants Vector b:")
print(b)
print("\nSolution (x, y, z):")
print(solution)
check = np.dot(A, solution)
print("\nVerification (A * solution):")
print(check)
```

1. Consider the following 10 points in X-Y plane: (−3, −7.6), (−2,1), (−1, 4.8), (0, 5), (1, 2.8), (2, −0.6), (3, −4), (4, −6.2), (5, −6), (7, 6.4). Plot these points in X-Y plane. Connect the points to guess the polynomial. Once you guess your polynomial, formulate a Linear System and then solve it to find the coefficient of the polynomial. Finally, check if the polynomial fits along with the above given points.

```python
import numpy as np
import matplotlib.pyplot as plt
points = [(-3, -7.6), (-2, 1), (-1, 4.8), (0, 5), (1, 2.8),
          (2, -0.6), (3, -4), (4, -6.2), (5, -6), (7, 6.4)]
x_vals = np.array([p[0] for p in points])
y_vals = np.array([p[1] for p in points])
plt.scatter(x_vals, y_vals, color='red', label='Given Points')
plt.plot(x_vals, y_vals, 'b--', alpha=0.5, label='Connecting Points')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Given Points in XY Plane")
plt.legend()
plt.grid(True, ls=':')
plt.show()
degree = 9
A = np.vander(x_vals, N=degree+1, increasing=True)
coeffs = np.linalg.solve(A, y_vals)
print("Polynomial Coefficients (lowest degree first):")
print(coeffs)
y_fit = np.polyval(coeffs[::-1], x_vals)
print("\nCheck if polynomial fits the points:")
print("Original y-values:", y_vals)
print("Fitted y-values:  ", y_fit)
x_dense = np.linspace(min(x_vals), max(x_vals), 200)
y_dense = np.polyval(coeffs[::-1], x_dense)
plt.scatter(x_vals, y_vals, color='red', label='Given Points')
plt.plot(x_dense, y_dense, color='green', label='Fitted Polynomial')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Fit to Given Points")
plt.legend()
plt.grid(True, ls=':')
plt.show()
```
